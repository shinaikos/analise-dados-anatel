Mostre as contagens de solicitações QtdeSolic por tipo de serviço Servico, em ordem decrescente pelas contagens

```python
list(db.minas.aggregate([
    {'$group': {'_id': { 'tipo' : "$Servico" }, 'count': {'$sum': 1}}},
    {'$match': {'_id': {'$ne': "QtdeSolic"}}},
    {'$sort': {'count': -1}}
]))
```
